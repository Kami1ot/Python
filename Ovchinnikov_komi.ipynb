```python
import random as ran
from itertools import permutations
def gen_cord(s):
  cords = []
  for i in range(s):
    x = ran.randint(0,100)
    y = ran.randint(0,100)
    cords.append((x,y))
  return cords

def dis(cord1,cord2):
  x1,y1 = cord1
  x2,y2 = cord2
  return round(((x1-x2)**2 + (y1-y2)**2)**0.5,2)

def ww(s):
  cords = gen_cord(s)
  print(cords)
  sw = 10000000
  qw = None
  for w in permutations(range(s)):
    w = list(w)
    w.append(w[0])
    z = len(w)
    m = 0
    for i in range(z-1):
      m+=dis(cords[w[i]],cords[w[i+1]])
      round(m,2)
      print(dis(cords[w[i]],cords[w[i+1]]),round(m,2))
    sw = min(sw,m)
    if (m == sw):
      qw = w
      print(w,round(sw,2))
    print()
  return qw,round(sw,2)

ww(5)
```
