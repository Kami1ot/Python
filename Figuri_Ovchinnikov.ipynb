```python
import matplotlib.pyplot as plt
import math
import random

plt.ylim(-20,20)
plt.xlim(-20,20)
ax = plt.gca()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.grid()

#задание 1,2

def pole(dat):      #простой вывод без всяких поворотов сдвигов и тд
  list(dat)
  for i in dat:
    i.append(i[0])
    plt.plot(list(zip(*i))[0],list(zip(*i))[1])

def sq(q=1):         #квадратик                sq(3) dat = cord = [[(x,y)(x,y)(x,y)(x,y)],[()()()()]]     plt.plot((x,y),(x,y))
  cord=[]
  x = 0
  for i in range(q):
    c= [(x,0),(x,1),(x+1,1),(x+1,0)]
    cord.append(c)
    x+=3
  return cord

def tg(q=1):        #треугольник
  cord=[]
  x = 0
  for i in range(q):
    c= [(x,0),(x + 0.5, (3 ** 0.5) / 2),(x+1,0)]
    cord.append(c)
    x+=3
  return cord

def hx(q=1):        #шестиугольник
  cord=[]
  x = 0
  for i in range(q):
    c= [(x, 0), (x + 0.5, (3 ** 0.5) / 2), (x + 1.5, (3 ** 0.5) / 2), (x + 2, 0), (x + 1.5, -(3 ** 0.5) / 2),(x + 0.5, -(3 ** 0.5) / 2)]
    cord.append(c)
    x+=3
  return cord

def four(n=1):
  x = 0
  for i in range(7):
    s = random.randint(1,3)
    if s == 1:
      pt(sq(),x)
      x+=2
    if s == 2:
      pt(tg(),x)
      x+=3
    else:
      pt(hx(),x)
      x+=3
    print(s)

pole(tg(4))
```

```python
plt.ylim(-10,10)
plt.xlim(-10,10)
ax = plt.gca()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.grid()


#задание 3
def pt_ex(cords,x=0.0,y=0.0):           #вложенная функция параллельного переноса которая используется в выводе оного переноса
  n_cords= []
  for cord in cords:
    n_cord=[]
    for c in cord:
      n_c = (c[0]+x,c[1]+y)
      n_cord.append(n_c)
    n_cords.append(n_cord)


  return n_cords

def pt(dat,x=0.0,y=0.0):              #по факту просто вывод параллельного переноса

  for cords in pt_ex(dat,x,y):
    cords.append(cords[0])
    plt.plot(list(zip(*cords))[0], list(zip(*cords))[1])

def tr_ex(cords,a=0):                #вложенная функция для вывода поворота
  n_cords = []
  for cord in cords:
      n_cord = []
      for c in cord:
        x,y = c
        n_c = (x * math.cos(math.radians(a)) - y * math.sin(math.radians(a)), x * math.sin(math.radians(a)) + y * math.cos(math.radians(a)))
        n_cord.append(n_c)
      n_cords.append(n_cord)
  return n_cords

def tr(dat,a=0):                    #вывод поворота
  for cords in tr_ex(dat,a):
    cords.append(cords[0])
    plt.plot(list(zip(*cords))[0], list(zip(*cords))[1])

def pt_tr(dat,a=0,x=0.0,y=0.0):     # вывод поворота и параллельного переноса одновременно
  for cords in pt_ex(tr_ex(dat,a),x,y):
    cords.append(cords[0])
    plt.plot(list(zip(*cords))[0], list(zip(*cords))[1])


def hom_ex (cords,k=1):             #вложенная функция для гомотетии(по факту как я понял гомотетия это пропорциональное увеличение фигуры)
  n_cords = []
  for cord in cords:
      n_cord = []
      for c in cord:
          n_c = (c[0] * k, c[1] * k)
          n_cord.append(n_c)
      n_cords.append(n_cord)

  return n_cords

def hom(dat,k=1):                   #вывод гомотетии
  for cords in hom_ex(dat,k):
    cords.append(cords[0])
    plt.plot(list(zip(*cords))[0], list(zip(*cords))[1])

def sim_ex(cords,k=1):
  n_cords = []
  for cord in cords:
    n_cord = []
    for c in cord:
      x,y = c
      n_c = (x, -y+k)
      n_cord.append(n_c)
    n_cords.append(n_cord)
  return n_cords

def sim(dat,k):
  for cords in sim_ex(dat,k):
    cords.append(cords[0])
    plt.plot(list(zip(*cords))[0], list(zip(*cords))[1])
  for cords in dat:
    cords.append(cords[0])
    plt.plot(list(zip(*cords))[0], list(zip(*cords))[1])

print(tg())
hom(tg(),10)
plt.show()

```
